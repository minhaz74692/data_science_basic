```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data.csv', parse_dates= ['day']) #Conver day column data type to date(TimeStamp)
```

```python
df
```

```python
# df.reset_index(inplace=True)
df.set_index('day', inplace=True)
```
