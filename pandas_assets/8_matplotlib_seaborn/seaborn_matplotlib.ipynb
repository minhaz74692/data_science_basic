```python
import pandas as pd
```

```python
ef = pd.ExcelFile('linechart.xlsx')
print(ef.sheet_names)
df = pd.read_excel('linechart.xlsx', sheet_name='product sales')
df
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(df['Quarter'], df['Fridge'], color = 'blue', label = 'Fridsge')
plt.plot(df['Quarter'], df['Dishwasher'], color = 'green', label = 'Dishwasher')
plt.plot(df['Quarter'], df['Washing Machine'], color = 'black', label = 'Washing Machine')
plt.title('Product Sales')
plt.xlabel('Financial Quarter')
plt.ylabel('Revenue(min $)', color = 'red', )
plt.legend()

plt.show()
```

```python
total_sales = df[['Fridge', 'Dishwasher', 'Washing Machine']].max()
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%',colors=['grey', 'yellow', 'green'], explode = (0.1, 0, 0), shadow=True, startangle= 90)
plt.show()
```
