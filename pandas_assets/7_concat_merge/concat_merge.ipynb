```python
import pandas as pd
```

```python
temperatue_df = pd.DataFrame({
    'city': ['chennai', 'delhi', 'mumbai'],
    'temperatue': [23,34,32]
}, index=[0,1,2])
temperatue_df
```

```python
windspeed_df = pd.DataFrame({
    'city': ['delhi', 'mumbai', 'kalkata'],
    'windspeed': [7,12, 23]
}, index = [1,2, 0])
windspeed_df
```

```python
pd.concat([temperatue_df, windspeed_df], axis=1)
```

```python
temperatue_df
```

```python
windspeed_df
```

```python
# it merge the common city in both table(Inner Join)
pd.merge(temperatue_df, windspeed_df, on = 'city', how = 'inner')
```

```python
# it merge all the row from left table and the common row from the right table(left Join)
pd.merge(temperatue_df, windspeed_df, on = 'city', how = 'left')
```

```python
# it merge all the row from right table and the common row from the left table(righe Join)
pd.merge(temperatue_df, windspeed_df, on = 'city', how = 'right')
```

```python
#Outer merge/join: merge all the row from both table
pd.merge(temperatue_df, windspeed_df, on = 'city', how = 'outer')
```
