```python
def standerdize_currency(curr):
    if curr == '$$' or curr == 'Dollars':
        return 'USD'
    return curr

df_financials_converted = pd.read_excel('movies_db.xlsx', 'financials', converters={
    'currency': standerdize_currency
    })

df_financials_converted.head(5)
```

# Marge Data Frame with a common row

```python
df_marged = pd.merge(df_movies, df_financials_converted, on='movie_id')
df_marged.head()
```

```python
df_marged.to_excel("marged_excel.xlsx", sheet_name='marged', index= False)
```

# Create Excel file with two sheets

```python
employee_df = pd.DataFrame({
    'name': ['Minhaz', 'Riyad', 'Rafid'],
    'age': [23,24,42],
    'designation': ['Flutter Developer', 'Python Developer', 'Android Developer']
})
employee_df
```

```python
department_df = pd.DataFrame({
    'name': ['Mobile', 'Web', 'Design'],
    'number': [2,3,1],
    'project': ['Help abode', 'PNI', 'Team Abode']
})
department_df
```

```python
with pd.ExcelWriter('new_created_office.xlsx') as writter:
    employee_df.to_excel(writter, sheet_name='emoloyee', index= False)
    department_df.to_excel(writter , sheet_name='department', index= False)
```
