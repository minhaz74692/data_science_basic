# DataFrame Using Python Dictionary

```python
weather_data = {
    'day': ['1/1/12', '2/1/12', '3/1/12'],
    'temp': [23,14,32],
    'windspeed': [6,2,7],
}

df = pd.DataFrame(weather_data)
df
```

# DataFrame Using python tupple

```python
weather_tupple_data = [
    ('1/1/23', 23.4, 'Rain'),
    ('2/1/23', 34.2, 'Sunny'),
    ('3/1/23', 34.2, 'Snow'),
]

df = pd.DataFrame(weather_tupple_data, columns= ['Day', 'Temperature', 'Events'])
df
```
